```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.__version__
```

```python
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.set(font_scale = 1.2)
sns.relplot(data=df, x = 'total_bill', y = 'tip', color = 'green', size = 'size', hue= 'size')
```

```python
df.describe(include=['category']).T
```

```python
sns.catplot(data= df, x ='day', y= 'total_bill')
```

```python
sns.catplot(data= df, x ='day', y= 'total_bill', kind ='box')
```

```python
sns.catplot(data= df, x ='day', y= 'total_bill', kind ='violin')
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.catplot(data=df, x = 'deck', kind = 'count', palette = 'Blues')
```
