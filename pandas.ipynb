```python
import pandas as pd
import numpy as np
```

```python
price = pd.Series(data = {'Apple' : 200, 'CD Project' : 60, 'Amazon' : 1900})
price
```

```python
price.count() # funkcja count zliczy ile elementów mamy w objekcie
```

```python
price.value_counts() #sprawdza rozkład wartości tzn np ile razy wystąpiła wartość 60
```

```python
price.describe() # pokazuje zbiorcze zestawiene
```

```python
price = pd.Series(data = {'Apple' : 200, 'CD Project' : 60, 'Amazon' : 1900, 'KGHM' : np.nan})
price
```

```python
price.nlargest(1)
```
